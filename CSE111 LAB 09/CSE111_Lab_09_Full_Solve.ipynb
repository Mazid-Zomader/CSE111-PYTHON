```python
class NikeBangladesh :
    branches = []
    stock = {'Air Jordan': 0, 'Cortez': 0, 'Zoom Kobe': 0}
    sold = 0
    def __init__ (self , name) :
        self.name = name
        NikeBangladesh.branches.append(self.name)
        self.individual_stock = {'Air Jordan': 0, 'Cortez': 0, 'Zoom Kobe': 0}
        self.individual_sold = 0
    @classmethod
    def status (cls) :
        print("Nike Bangladesh Status: ")
        print(f"Branches open: {cls.branches}")
        print("Currently Stocked: ")
        print(cls.stock)
        print(f"Sold: {cls.sold}")
    def details (self) :
        print(f"Nike {self.name} outlet:")
        print("Products Currently Stocked:")
        print(self.individual_stock)
        print(f"Sold: {self.individual_sold}")
    def restockProducts (self , dict) :
        for key , value in dict.items ():
            self.individual_stock[key] += dict[key]
            NikeBangladesh.stock[key] += dict[key]
    def productSold (self , dict) :
        for key in dict.keys() :
            self.individual_stock[key] -= dict[key]
            NikeBangladesh.stock[key] -= dict[key]
            NikeBangladesh.sold += dict[key]
#Driver Code
print("xxxxxxxxxxxxxx1xxxxxxxxxxxxxxxx")
NikeBangladesh.status()
dhaka = NikeBangladesh("Dhaka Banani")
chittagong = NikeBangladesh("Chittagong GEC")
print("xxxxxxxxxxxxxx2xxxxxxxxxxxxxxxx")
dhaka.details()
print("xxxxxxxxxxxxxx3xxxxxxxxxxxxxxxx")
chittagong.details()
print("xxxxxxxxxxxxxx4xxxxxxxxxxxxxxxx")
dhaka.restockProducts(
{"Air Jordan":1200,"Cortez":200,"Zoom Kobe":200})
chittagong.restockProducts(
{"Air Jordan":1000,"Cortez":250,"Zoom Kobe":100})
print("xxxxxxxxxxxxxx5xxxxxxxxxxxxxxxx")
NikeBangladesh.status()
print("xxxxxxxxxxxxxx6xxxxxxxxxxxxxxxx")
dhaka.productSold({"Air Jordan":760,"Cortez":90})
chittagong.productSold({"Air Jordan":520,"Zoom Kobe":70})
print("xxxxxxxxxxxxxx7xxxxxxxxxxxxxxxx")
NikeBangladesh.status()
```

```python
class SultansDine :
    no_of_branches = []
    total = 0
    individual_sell = []
    def __init__ (self , name) :
        self.name = name
        SultansDine.no_of_branches.append(self.name)
        self.quantity = 0
    def sellQuantity (self , quan) :
        self.quantity = quan
    def branchInformation (self) :
        print(f"Branch Name: {self.name}")
        Branch_sell = 0
        if self.quantity < 10:
            Branch_sell = self.quantity * 300
        elif self.quantity < 20:
            Branch_sell = self.quantity * 350
        else: 
            Branch_sell = self.quantity * 400
        print(f"Branch Sell: {Branch_sell} Taka")
        SultansDine.total += Branch_sell
        SultansDine.individual_sell.append(Branch_sell)
    @classmethod
    def details (cls) :
        print(f"Total Number of branch(s): {len(cls.no_of_branches)}")
        print(f"Total Sell: {cls.total} Taka")
        if len(cls.no_of_branches) == 0 :
            pass
        else :
            for index in range(len(cls.no_of_branches)) :
                print(f"Branch Name: {cls.no_of_branches[index]}, Branch Sell: {cls.individual_sell[index]} Taka")
                #percentage = (cls.individual_sell[index]/cls.total)*100
                print(f"Branach consists of total sell's: {round(((cls.individual_sell[index]/cls.total)*100),2)}%")
#Driver Code
SultansDine.details()
print('########################')
dhanmondi = SultansDine('Dhanmondi')
dhanmondi.sellQuantity(25)
dhanmondi.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()
```
