```python
#Task - 01
class PlayerEarning :
  def __init__ (self, name) :
    self.name = name
    self.salary = 0
    self.bonus = 0
  def calculateTotal(self, salary, goal=0) :
    self.salary = salary
    if goal > 30 :
      self.bonus = (5/100)*self.salary+10000
    elif goal == 0 :
      self.bonus = 0
    else :
      self.bonus = (5/100)*self.salary
  def printDetails(self) :
    print(f"Player Name: {self.name}")
    print(f"Player Season Earning without bonus: {self.salary}")
    print(f"Bonus = {int(self.bonus)}")
    print(f"Player Season Earning After Bonus: {int(self.bonus+self.salary)}")
print("**********************")
player1 = PlayerEarning('Buffon')
player1.calculateTotal(250000)
player1.printDetails()

print("\n**********************")
player2 = PlayerEarning('Dybala')
player2.calculateTotal(250000, 31)
player2.printDetails()

print("\n**********************")
player3 = PlayerEarning('Cuadrado')
player3.calculateTotal(250000, 20)
player3.printDetails()
```

```python
#Task - 02
class myList :
  def __init__ (self, *args) :
    self.list = list(args)
  def sum (self) :
    sum = 0
    for index in range(len(self.list)) :
      sum += self.list[index]
    print("Sum:",sum)
  def merge(self, *args) :
    for index in range(len(args)) :
      self.list.append(args[index])
  def average(self) :
    if len(self.list) == 0 :
      print("Average:",0)
    else :
      sum = 0
      for index in range(len(self.list)) :
        sum += self.list[index]
      print("Average:",sum/len(self.list))
l1 =  myList(2,3,4,5,6) #you might need a list inside your class to store the values
l1.sum()
l1.merge(4,5,9)
l1.sum()
l1.average()
print("-----------------------------")
l2 =  myList()
l2.average()
l2.merge(1,2,4,8)
l2.sum()
```

```python
#Task - 03
class Bird :
  def __init__(self, name, status= False) :
    self.name = name
    self.status = status
    self.__type = "Flightless Birds"
  def fly(self) :
    if self.status == False :
      print(f"{self.name} can not fly")
    else :
      print(f"{self.name} can fly")
  def setType(self, status) :
    self.__type = status
  def getType(self) :
    return self.__type
  def printDetail(self) :
    print(f"Name: {self.name}")
    print(f"Type: {self.getType()}")
ostrich = Bird('Ostrich')
duck = Bird("Duck", True)
owl = Bird('Owl', True)
print("###########################")
ostrich.fly()
duck.fly()
owl.fly()
duck.setType('Water Birds')
owl.setType('Birds of Prey')
print("=========================")
ostrich.printDetail()
print("=========================")
duck.printDetail()
print("=========================")
owl.printDetail()
```

```python
#Task - 04
class Account :
  count = 0
  def __init__(self, name, age, prof, salary) :
    self.name, self.age, self.profession, self.balance = name, age, prof, salary
    Account.count += 1
  def addMoney (self, value) :
    self.balance += value
  def withdrawMoney(self, value) :
    if self.balance - value >= 0 :
      self.balance -= value
    else :
      pass
  def printDetails(self) :
    print(f"Name: {self.name}")
    print(f"Age: {self.age}")
    print(f"Occupation: {self.profession}")
    print(f"Total Amount: {self.balance}")

# Write your code here

print('No of account holders:', Account.count)
print("=========================")
p1 = Account("Abdul", 45, "Service Holder", 500000)
p1.addMoney(300000)
p1.printDetails()
print("=========================")
p2 = Account("Rahim", 55, "Businessman", 700000)
p2.withdrawMoney(700000)
p2.printDetails()
print("=========================")
p3 = Account("Ashraf", 62, "Govt. Officer", 200000)
p3.withdrawMoney(250000)
p3.printDetails()
print("=========================")
print('No of account holders:', Account.count)
```

```python
#Task - 05
class Smartphone :
    def __init__(self, name = None) :
        self.name = name
        self.feature = {}
    def addFeature(self, parts, spec) :
        if self.name == None :
            print(f"Feature can not be added without phone name")
        else :
            if parts in self.feature.keys() :
                self.feature[parts].append(spec)
            else :
                self.feature[parts]=[]
                self.feature[parts].append(spec)
    def setName (self, name) :
        self.name = name
    def printDetail(self) :
        print(f"Phone Name: {self.name}")
        for key, value in self.feature.items() :
            print(f"{key}: {','.join(value)}")
s1 = Smartphone()
print("=================================")
s1.addFeature("Display", "6.1 inch")
print("=================================")
s1.setName("Samsung Note 20")
s1.addFeature("Display", "6.1 inch")
s1.printDetail()
print("=================================")
s2 = Smartphone("Iphone 12 Pro")
s2.addFeature("Display", "6.2 inch")
s2.addFeature("Ram", "6 GB")
print("=================================")
s2.printDetail()
s2.addFeature("Display", "Amoled panel")
s2.addFeature("Ram", "DDR5")
print("=================================")
s2.printDetail()
print("=================================")
```

```python
#Task - 10
class Student:
    def __init__(self,name,ID):
        self.name = name
        self.ID = ID
    def Details(self):
        return "Name: "+self.name+"\n"+"ID: "+self.ID+"\n"
#Write your code here
class CSEStudent(Student):
    def __init__(self,name, id, semester) :
        super().__init__(name, id) 
        self.semester = semester
        self.course = {}
    def Details(self) :
        return f"{super().Details()}Current Semester: {self.semester}"
    def addCourseWithMarks(self, *args ) :
        for index in range(0, len(args), 2 ) :
            if args[index+1] >= 85 :
                self.course[args[index]] = 4.0
            elif args[index+1] >= 80 :
                self.course[args[index]] = 3.3
            elif args[index+1] >= 70 :
                self.course[args[index]] = 3.0
            elif args[index+1] >= 65 :
                self.course[args[index]] = 2.3
            elif args[index+1] >= 57 :
                self.course[args[index]] = 2.0
            elif args[index+1] >= 55 :
                self.course[args[index]] = 1.3
            elif args[index+1] >= 50 :
                self.course[args[index]] = 1.0
            else :
                self.course[args[index]] = 0.0
    def showGPA(self) :
        gpa = 0
        for key, value in self.course.items() :
            gpa += value
            print(f"{key}: {value}")
        gpa /= len(self.course.values())
        print(f"GPA of{self.name} is: {round(gpa, 2)}")
    
                
Bob = CSEStudent("Bob","20301018","Fall 2020")
Carol = CSEStudent("Carol","16301814","Fall 2020")
Anny = CSEStudent("Anny","18201234","Fall 2020")
print("#########################")
print(Bob.Details())
print("#########################")
print(Carol.Details())
print("#########################")
print(Anny.Details())
print("#########################")
Bob.addCourseWithMarks("CSE111",83.5,"CSE230",73.0,"CSE260",92.5)
Carol.addCourseWithMarks("CSE470",62.5,"CSE422",69.0,"CSE460",76.5,"CSE461",87.0)
Anny.addCourseWithMarks("CSE340",45.5,"CSE321",95.0,"CSE370",91.0)
print("----------------------------")
Bob.showGPA()
print("----------------------------")
Carol.showGPA()
print("----------------------------")
Anny.showGPA()
```
